```python
from openai import OpenAI
import json

KEY = load(open('./key.json', 'r'))
client = OpenAI(api_key=KEY['OpenAIKey'])

def show_map(light_pollution, near_trail, near_road, near_river, mobile_reception):
    """Shows campsites along with layers that the user wanted to show"""
    
    
    pass


# Dummy campsite recommendation function, hardcoded for demonstration
def recommend_campsites(light_pollution, near_trail, near_road, near_river, mobile_reception):
    """Recommend campsites based on individual weighted preferences"""
    # Ensure weights sum to 1
    total_weight = light_pollution + near_trail + near_road + near_river + mobile_reception
    if abs(total_weight - 1) > 0.01:  # Allowing a small margin for floating-point arithmetic
        return json.dumps({"error": "Weights do not sum to 1"})

    weights = {
        "light_pollution": light_pollution,
        "near_trail": near_trail,
        "near_road": near_road,
        "near_river": near_river,
        "mobile_reception": mobile_recept}
    print(weights)

    # Example dummy response
    
    return json.dumps({
        "recommendations": [
            {"name": "Camp Elsie", "match_score": "90%"},
            {"name": "Campsite HouWIe", "match_score": "55%"}
        ],
        "weights_used": weights
    })

def run_campsite_recommendation():
    messages = []
    tools = [
        {
            "type": "function",
            "function": {
                "name": "recommend_campsites",
                "description": "Recommend campsites based on weighted preferences. \
                                Weights for each should be provided by your estimate of what the user \
                                may want based on their preferences and sum to 1.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "light_pollution": {
                            "type": "number",
                            "description": "Weight for low light pollution, important for stargazing"
                        },
                        "near_trail": {
                            "type": "number",
                            "description": "Weight for proximity to trails"
                        },
                        "near_road": {
                            "type": "number",
                            "description": "Weight for proximity to roads"
                        },
                        "near_river": {
                            "type": "number",
                            "description": "Weight for proximity to rivers"
                        },
                        "mobile_reception": {
                            "type": "number",
                            "description": "Weight for good mobile reception"
                        }
                    },
                    "required": ["light_pollution", "near_trail", "near_road", "near_river", "mobile_reception"],
                },
            },
        }
    ]

    # Wait for user input
    user_message = input("Please describe your ideal camping experience: ")
    messages.append({
        "role": "system",
        "content": "I'm here to help you find the campsite based on your preferences. My job is to run a recommending function based on weights, that I'll interpret from your preferences and feed into the function."
    },)
    messages.append({"role": "user", "content": user_message})

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        tools=tools,
        tool_choice="auto",
    )

    response_message = response.choices[0].message
    tool_calls = response_message.tool_calls

    if tool_calls:
        print("called")
        available_functions = {"recommend_campsites": recommend_campsites}
        messages.append(response_message)  # Add the assistant's reply to the conversation

        for tool_call in tool_calls:
            function_name = tool_call.function.name
            function_to_call = available_functions[function_name]

            function_args = json.loads(tool_call.function.arguments)
            light_pollution = function_args.get("light_pollution", 0)
            near_trail = function_args.get("near_trail", 0)
            near_road = function_args.get("near_road", 0)
            near_river = function_args.get("near_river", 0)
            mobile_reception = function_args.get("mobile_reception", 0)

            recommendations = function_to_call(
                light_pollution=light_pollution,
                near_trail=near_trail,
                near_road=near_road,
                near_river=near_river,
                mobile_reception=mobile_reception
            )

            messages.append(
                {
                    "tool_call_id": tool_call.id,
                    "role": "tool",
                    "name": function_name,
                    "content": recommendations,
                }
            )

        final_response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages,
        )
        return final_response
```

```python
print(run_campsite_recommendation())
```
